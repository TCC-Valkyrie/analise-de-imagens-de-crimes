```python
# Deep Learning Libraries
import torch
from ultralytics import YOLO
```

```python
# Testing CUDA availability
print("CUDA Available: ", torch.cuda.is_available())
print("GPU Name: ", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "None")
```

```python
# 0. Set random seed for reproducibility
torch.manual_seed(42)

# 1. Load a pre-trained YOLOv8 nano model
model = YOLO("yolov8m.pt") 


if torch.cuda.is_available():
    model.to("cuda")
    
# 2. Train the model on your dataset
results = model.train(
    data="dataset/data.yaml",  # Path to your YAML configuration file
    epochs=100,                # Number of times the model will see the entire dataset
    imgsz=640,                 # Standard image size for YOLO
    batch=16,                  # Number of images processed in one batch (lower if you run out of memory)
    project="robbery_model",   # Folder where your results will be saved
    name="initial_training"    # Subfolder for this specific training run
)
```
